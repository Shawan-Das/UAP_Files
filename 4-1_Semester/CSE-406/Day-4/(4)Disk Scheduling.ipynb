### C-SCAN Disk Scheduling Algorithm

```python
#Time Input
data=list(map(int, input("Enter the value for queue: ").split(", ")))

#Disk Head
head=int(input("Disk Head: "))
#Upper head/ lower head
upper_head=int(input("Upper Head: "))
lower_head=int(input("Lower Head: "))

#Upper level data/ lower level data of Disk Head.
upper_list=[i for i in data if i>=head]     # store the valuse larger than head
lower_list=[i for i in data if i<head]      # store the valuse smaller than head


if head < data[0]:
  seeking_time = abs(upper_head - head) + abs(upper_head- lower_head) + abs(lower_head - max(lower_list))
  # go to Right morst (Upper Bound) position the directly back to Lower bound and then complete rest of the queue values
else:
  seeking_time = abs(lower_head - head) + abs(lower_head - upper_head) + abs(upper_head - max(upper_list))
  # got to Left most (Lower Bound) position then directly jump to Upper bound then perform rest of the queue values

print("Seeking Time : ", seeking_time)
```
