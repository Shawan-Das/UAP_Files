### Shortest Job First

```python
## Number of Process
n= int(input("Total Process: "))

#process Arrival time & Burst Time
process=[]
Arrival_time=[]
Burst_time=[]

## input for Arrival-Burst Time
for i in range(n):
  at, bt = input().split(" ")  # split input data by space
  process.append(i+1)          # process id
  Arrival_time.append(int(at)) # insret Arrival time in list
  Burst_time.append(int(bt))   # insert Burst time in list


# sort all data at the basis of Burst_time
Burst_time,Arrival_time,process = zip(*sorted(zip(Burst_time,
                                                  Arrival_time,
                                                  process)))

## Calculate Turn Around time/ Waiting Time/ Complition Time

Completion_time=[] 
Turn_around_time=[]
Waiting_time=[]

for i in range(n):
  if(i==0):
    Completion_time.append(0+Burst_time[i]) # Copletion time for process-1 will be the burst time
  else:
    Completion_time.append( Completion_time[i-1] + Burst_time[i] ) # for other process, Completion time - = Prvious Completion time + BUrst time
  
  Turn_around_time.append(Completion_time[i]-Arrival_time[i]) # Turn around time = Completion Time - Arrival Time
  if(i==0):
    Waiting_time.append(0) # No waiting time for process-1
  else:
    Waiting_time.append( Turn_around_time[i] - Burst_time[i] ) # Waiting time = Turn Around time - Burst time


# sort everything at the basis of Process id
process, Arrival_time, Burst_time, Completion_time,Turn_around_time,Waiting_time= zip(*sorted(zip( process, 
                                                                                                   Arrival_time,
                                                                                                   Burst_time,
                                                                                                   Completion_time,
                                                                                                   Turn_around_time,
                                                                                                   Waiting_time )))

print(Arrival_time)
print(Burst_time)
print(Completion_time)
print(Turn_around_time)
print(Waiting_time)

average_time=(sum(Waiting_time)/n);  # Average time = sum of Waiting time / total process
print("Average Waiting Time: ",average_time)
```
