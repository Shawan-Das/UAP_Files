### First Come First Serve

```python
## Number of Process
n= int(input("Total Process: "))

#process Arrival time & Burst Time
Arrival_time=[]
Burst_time=[]

## input for Arrival-Burst Time
for i in range(n):
  at, bt = input().split(" ")
  Arrival_time.append(int(at))
  Burst_time.append(int(bt))

## Calculate Turn Around time/ Waiting Time/ Complition Time

Completion_time=[]
Turn_around_time=[]
Waiting_time=[]

for i in range(n):
  if(i==0):
    Completion_time.append(0+Burst_time[i])
  else:
    Completion_time.append( Completion_time[i-1] + Burst_time[i] )
  
  Turn_around_time.append(Completion_time[i]-Arrival_time[i])
  if(i==0):
    Waiting_time.append(0)
  else:
    Waiting_time.append( Turn_around_time[i] - Burst_time[i] )
  

print(Arrival_time)
print(Burst_time)
print(Completion_time)
print(Turn_around_time)
print(Waiting_time)

average_time=(sum(Waiting_time)/n);
print("Average Waiting Time: ",average_time)



```
