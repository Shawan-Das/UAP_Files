### FCFS Disk Scheduling Algorithm

image.png

```python
#Time Input
data=list(map(int, input("Enter the value for queue: ").split(", ")))

#Disk Head
head=int(input("Disk Head: "))

#sub_list used for storing individual Seeking time
data2=[] # used for calculating Seeking time.

data2.append(abs(data[0]-head)) # first queue's seeking time is abs of first data in queue - Head(Current Position at the start of process)
for i in range(1,len(data)):
  data2.append(abs(data[i]-data[i-1]))  # current queue's seeking time is, abs of queue value - Current position of The Disk

print("Seek Time: ",sum(data2))  # Total seeking time= sum of All queue's seeking time which is in Data2

```

### Shortest Seek Time First(SSTF)

image.png

```python
#Time Input
data=list(map(int, input("Enter the value for queue: ").split(", ")))

#Disk Head
head=int(input("Disk Head: "))

# List for storing individual Seeking time
data2=[]

while(data!=[]):
  temp=[abs(data[i]-head) for i in range(len(data))]  #store the difrence between Head and Current values of list
  flag=min(temp)                                      # find the lowest value to get lowest distance
  flag=temp.index(flag)                               # distance index
  data2.append(temp[flag])                            # inset seeking time to new list
  head=data[flag]                                     # Change the Head to new position
  temp.pop(flag)                                      # delete completed data
  data.pop(flag)                                      # delete Completed task
  #print(data) 
print("Total Seeking Time: ",sum(data2))              # get Seeking time
```

### SCAN (Elevator) Disk Scheduling Algorithms
image.png

```python
#Time Input
data=list(map(int, input("Enter the value for queue: ").split(", ")))

#Disk Head
head=int(input("Disk Head: "))
#Upper head/ lower head
upper_head=int(input("Upper Head: "))
lower_head=int(input("Lower Head: "))


if head < data[0]:
  seeking_time= abs((upper_head-head)) + abs(upper_head-min(data))  # go to right most then back to lowest data in queue
else:
    seeking_time= abs(lower_head- head) + abs(max(data)-lower_head) # go to left most then come back to max data in queue
print("Seeking Time : ",seeking_time)
```

### C-SCAN Disk Scheduling Algorithm
image.png

```python
#Time Input
data=list(map(int, input("Enter the value for queue: ").split(", ")))

#Disk Head
head=int(input("Disk Head: "))
#Upper head/ lower head
upper_head=int(input("Upper Head: "))
lower_head=int(input("Lower Head: "))

#Upper level data/ lower level data of Disk Head.
upper_list=[i for i in data if i>=head]     # store the valuse larger than head
lower_list=[i for i in data if i<head]      # store the valuse smaller than head


if head < data[0]:
  seeking_time = abs(upper_head - head) + abs(upper_head- lower_head) + abs(lower_head - max(lower_list))
  # go to Right morst (Upper Bound) position the directly back to Lower bound and then complete rest of the queue values
else:
  seeking_time = abs(lower_head - head) + abs(lower_head - upper_head) + abs(upper_head - max(upper_list))
  # got to Left most (Lower Bound) position then directly jump to Upper bound then perform rest of the queue values

print("Seeking Time : ", seeking_time)
```

### LOOK Disk Scheduling Algorithm

image.png

```python
#Time Input
data=list(map(int, input("Enter the value for queue: ").split(", ")))

#Disk Head
head=int(input("Disk Head: "))

if head < data[0]:
  # visit right top most max value of the queue, then get back to most min value of the queue.
  seeking_time= abs( head - max(data)) + abs(max(data) - min(data))
else:
  # visit right top most min value of the queue, then get back to most max value of the queue.
  seeking_time= abs( head - min(data)) + abs(min(data) - max(data))

print("Total Seeking time: ",seeking_time)
```

### C-LOOK Disk Scheduling Algorithm

image.png

```python
#Time Input
data=list(map(int, input("Enter the value for queue: ").split(", ")))

#Disk Head
head=int(input("Disk Head: "))

#Upper level data/ lower level data of Disk Head.
upper_list=[i for i in data if i>=head]     # store the valuse larger than head
lower_list=[i for i in data if i<head]      # store the valuse smaller than head


if head < data[0]:
  seeking_time = abs(max(upper_list) - head) + abs(max(upper_list)- min(lower_list)) + abs(min(lower_list) - max(lower_list))
  # go to Max value of the Queue then directly jump to Lowest vaue of the Queue then complete rest of the queue values
else:
  seeking_time = abs(min(lower_list) - head) + abs(min(lower_list) - max(upper_list)) + abs(max(upper_list) - min(upper_list))
  # go to Min value of the Queue then directly jump to Max vaue of the Queue then complete rest of the queue values

print("Seeking Time : ", seeking_time)
```
