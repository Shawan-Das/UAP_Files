```python
from typing_extensions import Text
dataset= pd.read_csv('/content/drive/MyDrive/4-1_Semester/CSE-404/Testing/cod.csv')
dataset.head()
list_drop = ['name','kills','deaths']
dataset.drop(list_drop, axis=1, inplace=True)
dataset.head()
```

```python
dataset.info()
```

```python
dataset.wins = dataset.wins.apply(lambda x: (x-min(dataset.wins))/(max(dataset.wins)-min(dataset.wins)))
dataset.kdRatio = dataset.kdRatio.apply(lambda x: (x-min(dataset.kdRatio))/(max(dataset.kdRatio)-min(dataset.kdRatio)))
dataset.killstreak = dataset.killstreak.apply(lambda x: (x-min(dataset.killstreak))/(max(dataset.killstreak)-min(dataset.killstreak)))
dataset.level = dataset.level.apply(lambda x: (x-min(dataset.level))/(max(dataset.level)-min(dataset.level)))
dataset.losses = dataset.losses.apply(lambda x: (x-min(dataset.losses))/(max(dataset.losses)-min(dataset.losses)))
dataset.prestige = dataset.prestige.apply(lambda x: (x-min(dataset.prestige))/(max(dataset.prestige)-min(dataset.prestige)))
dataset.hits = dataset.hits.apply(lambda x: (x-min(dataset.hits))/(max(dataset.hits)-min(dataset.hits)))
dataset.timePlayed = dataset.timePlayed.apply(lambda x: (x-min(dataset.timePlayed))/(max(dataset.timePlayed)-min(dataset.timePlayed)))
dataset.averageTime = dataset.averageTime.apply(lambda x: (x-min(dataset.averageTime))/(max(dataset.averageTime)-min(dataset.averageTime)))
dataset.gamesPlayed = dataset.gamesPlayed.apply(lambda x: (x-min(dataset.gamesPlayed))/(max(dataset.gamesPlayed)-min(dataset.gamesPlayed)))
dataset.assists = dataset.assists.apply(lambda x: (x-min(dataset.assists))/(max(dataset.assists)-min(dataset.assists)))
dataset.misses = dataset.misses.apply(lambda x: (x-min(dataset.misses))/(max(dataset.misses)-min(dataset.misses)))
dataset.xp = dataset.xp.apply(lambda x: (x-min(dataset.xp))/(max(dataset.xp)-min(dataset.xp)))
dataset.scorePerMinute = dataset.scorePerMinute.apply(lambda x: (x-min(dataset.scorePerMinute))/(max(dataset.scorePerMinute)-min(dataset.scorePerMinute)))
dataset.shots = dataset.shots.apply(lambda x: (x-min(dataset.shots))/(max(dataset.shots)-min(dataset.shots)))
dataset.headshots = dataset.headshots.apply(lambda x: (x-min(dataset.headshots))/(max(dataset.headshots)-min(dataset.headshots)))
dataset.head()
```

```python
dataset.plot(figsize=(40,10))
```

```python
x= dataset[['wins','kdRatio','killstreak','level','losses','prestige','hits','timePlayed','averageTime','gamesPlayed','assists','misses','xp','scorePerMinute','shots']].values
y= dataset['headshots'].values

print(x.shape, y.shape)
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(x,y, test_size=0.2, random_state=1)

print(X_train.shape, y_train.shape)
print(X_test.shape, y_test.shape)
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(X_train,y_train)
print(model.coef_)
print('----------')
print("Model Accuracy: ",model.score(X_test, y_test))
```

```python
i=1
while True:
  X_train,X_test,y_train,y_test=train_test_split(x,y, test_size=0.2)
  model.fit(X_train, y_train)
  print(i, end=" ")
  i+=1
  print("----------",model.score(X_test, y_test))
  if model.score(X_test,y_test)>0.98:
    break

# Max got 38076/ 0.68
```

```python
## setting plot style
plt.style.use('fivethirtyeight')
  
## plotting residual errors in training data
plt.scatter(model.predict(X_train), model.predict(X_train) - y_train,
            color = "green", s = 10, label = 'Train data')
  
## plotting residual errors in test data
plt.scatter(model.predict(X_test), model.predict(X_test) - y_test,
            color = "blue", s = 10, label = 'Test data')


## plotting line for zero residual error
#plt.hlines(y = 0, xmin = 0, xmax = 2.1, linewidth = 2)
  
## plotting legend
plt.legend(loc = 'upper right')
  
## plot title
plt.title("Residual errors")

plt.figure(figsize=(12,6))

## method call for showing the plot
plt.show()
```
