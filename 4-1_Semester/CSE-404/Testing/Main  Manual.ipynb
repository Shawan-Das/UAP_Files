```python
from typing_extensions import Text
dataset= pd.read_csv('/content/drive/MyDrive/4-1_Semester/CSE-404/Testing/cod.csv')
dataset.head()
list_drop = ['name','kills','deaths']
dataset.drop(list_drop, axis=1, inplace=True)
dataset.head()
```
