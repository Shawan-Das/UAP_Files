```python
brust_time = []
num_process = int(input("Enter the number of Processes: "))
process_seq = 1

for i in range(num_process):
    print("For process P", process_seq, end='')
    brust_time.append(int(input(", Burst Time = ")))
    process_seq = process_seq + 1
print("\n")

wait_time = []
avg_wait_time = 0
wait_time.insert(0, 0)

for j in range(1, len(brust_time)):
    wait_time.insert(j, (wait_time[j - 1] + brust_time[j - 1]))

avg_wait_time += wait_time[j]
avg_wait_time = float(avg_wait_time / num_process)
process = 1
for k in range(0, num_process):
    print(wait_time[k], "P", process, brust_time[k], end=' | ')
    process += 1
print("\n Average Waiting Time = ", avg_wait_time)
```

```python
def A(m,n):
  if m==0:
    return n+1
  elif n==0:
    return A(m-1,1)
  else:
    return A(m-1,A(m,n-1))
```

```python
print(A(3,6))
```
