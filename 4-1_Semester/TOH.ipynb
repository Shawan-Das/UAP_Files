```python
def TOH(disk, s, m, d):
  if(disk==1):
    print("Move disk 1 from ",s," to ",d)
  else:
    TOH(disk-1,s,d,m)
    print("Move ",disk," from ",s," to ",d)
    TOH(disk-1,m,s,d)
```

```python
number= int(input("Number of Disk: "))
source= input("Source: ")
medium= input("Medium: ")
dest = input("Destination: ")
TOH(number,source,medium,dest)
```
