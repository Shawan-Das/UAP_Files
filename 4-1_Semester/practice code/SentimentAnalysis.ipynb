```python
#Inport Liraries
from textblob import TextBlob
import nltk
from newspaper import Article
```

```python
#Get Article
# https://www.english-for-students.com/Very-Funny-Short-Stories.html

url = 'https://www.english-for-students.com/Foolish-Imitation.html'
article = Article(url)
```

```python
# NLP - [download]
article.download()
article.parse()
nltk.download('punkt')
article.nlp()
```

```python
#Text summary of the article

text = article.summary
print(text)
```

```python
#Create a textblob object
# Analys sentiment
obj= TextBlob(text)

# return a value between -1 to 1
# < 0 [negative]
# 0 [netural]
# > 0 [positive]
#
sentiment = obj.sentiment.polarity
print(sentiment)
```
