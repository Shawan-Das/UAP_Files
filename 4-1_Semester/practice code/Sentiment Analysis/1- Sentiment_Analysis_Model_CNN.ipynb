### Processing Data

```python
stop_words = set(stopwords.words("english"))
lemmatizer = WordNetLemmatizer()
```

```python
def clean_text(text):
  text = re.sub(r'[^\w\s]','',text, re.UNICODE)
  text = re.sub('[^a-zA-Z]', ' ', text)
  text = re.sub(r"\s+[a-zA-Z]\s+", ' ', text)
  text = re.sub(r'\s+', ' ', text)
  text = [lemmatizer.lemmatize(token) for token in text.split(" ")]
  text = [lemmatizer.lemmatize(token, "v") for token in text]
  text = [word for word in text if not word in stop_words]
  text = " ".join(text)
  return text
```

```python
dataset['Processed_Reviews'] = dataset.review.apply(lambda x: clean_text(x))
```

### Adding CNN & Maxpooling Layer to the Model

```python
embedding_dim = 50
maxlen =200
X_train, X_test, y_train, y_test = train_test_split(dataset['Processed_Reviews'], dataset['review'], test_size= 0.15, random_state= 0)
```

```python
dataset.sentiment.unique()
```

```python
X_train = tokenizer.texts_to_sequences(X_train)
X_test = tokenizer.texts_to_sequences(X_test)

X_train = pad_sequences(X_train, maxlen=maxlen, truncating='post', padding='post')
X_test = pad_sequences(X_test, maxlen=maxlen, truncating='post', padding='post')
```

```python
print(X_train.shape, y_train.shape)
print(X_test.shape, y_test.shape)
```

```python
vcoad_size = len(tokenizer.word_index) + 1
embedding_vecor_length = 32
callback = EarlyStopping (monitor = 'val_loss', patience=2)
```

```python
model_CNN = Sequential()
model_CNN. add (Embedding(input_dim = vocab_size, output_dim = embedding_dim, input_length = maxlen))
model_CNN. add (Conv1D(filters=32, kernel_size= 8,padding ='same', activation= 'relu'))
model_CNN. add (MaxPooling1D(pool_size = 2))
model_CNN. add (Dropout(0.5))

model_CNN. add (Conv1D(filters=32, kernel_size=3, padding = 'same',activation = 'relu'))
model_CNN. add (MaxPooling1D(pool_size = 2))
model_CNN. add (Conv1D(filters=32, kernel_size=3, padding = 'same',activation = 'relu'))
model_CNN. add (MaxPooling1D(pool_size = 2))
model_CNN. add (Dense(1, activation= 'sigmoid'))
#model_CNN. add (GlobalMaxPool1D())

model_CNN.compile(loss='sparse_categorical_crossentropy', optimizer = 'adam', metrics = ['accuracy'])
```

```python
model_CNN.summary()
```

### Model Fit

```python
history = model_CNN.fit(X_train, y_train, epochs=10, batch_size=128, verbose=1, validation_data= (X_test, y_test))
```
